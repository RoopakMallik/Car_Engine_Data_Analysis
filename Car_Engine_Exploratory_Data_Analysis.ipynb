```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from sklearn.linear_model import LinearRegression
```

```python
main_df=pd.read_csv('Car EU Data.csv')
main_df
```

```python
main_df.info()
```

```python
main_df.describe
```

```python
# checking for the null values
main_df.isnull().sum()
```

```python
# finding the car names
auto_names=main_df['name']
print(auto_names)
```

```python
#finding the maximum horespower
max_hp=main_df['horsepower'].max()
max_horsepower=main_df[main_df['horsepower']==max_hp]
max_horsepower
```

```python
#finding the minimum horespower
min_hp=main_df['horsepower'].min()
min_horsepower=main_df[main_df['horsepower']==min_hp]
min_horsepower
```

```python
# finding the average horespower
avg_hp=main_df['horsepower'].mean()
print('The Average Car Horsepower is:',avg_hp)
```

```python
#finding the median horespower
median_hp=main_df['horsepower'].median()
print('The Median Car Horsepower is:',median_hp)
```

```python
# finding the cars with maximum cylinders 
max_cylinders=main_df['cylinders'].max()
max_cylinders=main_df[main_df['cylinders']==max_cylinders]
print('Cars with the Most Cylinders are :',max_cylinders)
```

```python
# finding the range of horespower
range_hp=max_hp-min_hp
print('The Range of Horsepower is:',range_hp)
```

```python
# finding the maximum acceleration
max_acceleration=main_df['acceleration'].max()
max_acceleration=main_df[main_df['acceleration']==max_acceleration]
max_acceleration
```

```python
# finding the minimum acceleration
min_acceleration=main_df['acceleration'].min()
min_acceleration=main_df[main_df['acceleration']==min_acceleration]
min_acceleration
```

```python
# finding the average acceleration
average_acceleration=main_df['acceleration'].mean()
print('The Average Acceleration of Cars is:',average_acceleration)
```

```python
# finding the variance of acceleration and horsepower
acceleration_var=main_df['acceleration'].var()
horsepower_var=main_df['horsepower'].var()

print('The Variance of Acceleration is:',acceleration_var)
print('The Variance of Horsepower is:',horsepower_var)
```

```python
# finding the covariance of acceleration and horsepower
covariance = main_df[['acceleration', 'horsepower']].cov().iloc[0, 1]
print("Covariance between Acceleration and Horsepower:", covariance)
```

```python
# line plot of acceleration vs horsepower

plt.figure(figsize=(10, 6))
plt.plot(main_df['horsepower'], main_df['acceleration'], marker='o')
plt.title('Line Plot of Acceleration vs Horsepower')
plt.xlabel('horsepower')
plt.ylabel('acceleration')
plt.show()
```

```python
# hypothesis testing

contingency_table = pd.crosstab(main_df['origin'], main_df['cylinders'])
```

```python
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

print("Chi-Square Test Statistic:", chi2)
print("p-value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies Table:")
print(expected)
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Contingency Table to show Origin vs Cylinders')
plt.xlabel('Cylinders')
plt.ylabel('Origin')
plt.show()
```

```python
# fuel distribution of cars from different countries

japan_cars = main_df[main_df['origin'] == 'Japan']['fuel']
us_cars = main_df[main_df['origin'] == 'US']['fuel']
```

```python
t_stat, p_value = stats.ttest_ind(japan_cars, us_cars)

print("T-Test Statistic:", t_stat)
print("p-value:", p_value)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(japan_cars, kde=True, label='Japan', color='green')
sns.histplot(us_cars, kde=True, label='US', color='orange')
plt.title('Fuel Efficiency Distribution: Japan vs United States')
plt.xlabel('Fuel Efficiency')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# histogram of acceleration in cars

plt.figure(figsize=(10, 6))
sns.histplot(main_df['acceleration'], kde=True, bins=30, color='red')
plt.title('Histogram of Acceleration in Cars')
plt.xlabel('Acceleration')
plt.ylabel('Frequency')
plt.show()
```

```python
# linear regression of horsepower vs acceleration


X = main_df[['horsepower']].dropna()  
y = main_df['acceleration'].dropna()  

X = X.loc[y.index]
model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)

plt.figure(figsize=(10, 6))
sns.scatterplot(x='horsepower', y='acceleration', data=main_df, label='Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.title('Linear Regression: Horsepower vs Acceleration')
plt.xlabel('Horsepower')
plt.ylabel('Acceleration')
plt.legend()
plt.show()
```

```python
# Printing the coefficients
print("Intercept:", model.intercept_)
print("Coefficient:", model.coef_[0])
```
